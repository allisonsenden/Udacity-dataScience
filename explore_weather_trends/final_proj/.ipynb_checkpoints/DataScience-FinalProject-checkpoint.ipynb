# DataSet Analyzed: No-Show Appointments

## Question Posed: What factors are more likely to affect the likelihood of a patient showing up for their medical appointment?

## Data Investigation/Wrangling Completed:
#### In order to make sure the data I was using was normalized and going to give me accurate results, I started by ensuring I took care of any empty values so that they did not skew the results. Seems that all the values were indeed filled in, so I was able to begin. Then, I analyzed what type of data populated each column to understand what data I had to work with.  The time of the appointment day was not recorded so I began by just analyzing the difference amount ahead of time the patient would schedule their appointment. In order to do so, I needed to change the ScheduledDay and AppointmentDay data types from objects to datetime data types. From there I was able to calculate the lead time, or the number of days ahead of time a patient scheduled their appointment. It seems like the majority of patients scheduled the same day and they had a higher rate of showing up to their appointment. Furthermore, the chart generated seemed to show that the appointments that were scheduled further out were more likely to have a patient not show up as opposed to the alternative. I did some analysis on the Neighbourhoods, although we were not given the facilities location, so I could not draw any conclusions from that data. Next, I found that the number  of appoinments made by females was double that of men. This also correlated to their no-show rate being double as we would expect since there is a larger sample size for that sex. Finally, the age of the patients was looked at. I found that the younger that patient was, the more likely they were to be a no-show to their appointment. This was true for patients 39 years of age or younger.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv('/Users/allisonsenden/Downloads/noshowappointments-kagglev2-may-2016.csv')

data.head()
```

```python
print(data.info())
#view data types and discover more about the data to be worked with
```

```python
data.rename(columns={'No-show': 'Noshow'}, inplace=True)
#dont want to run into times when using a hypenated columm name will have troubles running the function I want
```

```python
data['Noshow'].value_counts()
#No means they showed for their appointment
#Yes means they missed their appointment
```

```python
noshow_patients = data[data.Noshow == 'Yes']
print(len(noshow_patients))
```

```python
total_patients = sum(data['Noshow'].value_counts())
```

```python
percent_noshow = len(noshow_patients) / total_patients
print(str(round(percent_noshow*100, 2)) + '% of total patients were no-shows to their appointments')
```

```python
data.ScheduledDay = pd.to_datetime(data.ScheduledDay)
data.AppointmentDay = pd.to_datetime(data.AppointmentDay)
#changing the data type of the appointment days to datetime instead of object
```

```python
def days(datetime):
    '''this function extracts the day out of the date/time stamp'''
    return datetime.date()
```

```python
data['lead_time'] = (
    data.AppointmentDay.apply(days)-data.ScheduledDay.apply(days)
)

#subtracting the appointment date from when the appointment was scheduled to get how many days prior the appointment
#was scheduled
```

```python
data.lead_time = ((data.lead_time.astype('timedelta64[D]')).astype(int))
#changing the lead_time days column to an integer data type instead of a datetime datatype
```

```python
leadtime_counts = data.lead_time.value_counts().sort_index()
print(leadtime_counts)
```

```python
grouped_lt = data.groupby(['lead_time', 'Noshow']).count()['PatientId']
grouped_lt

#grouped the lead time and the no show rate for that lead time together to see how the no show rate
#is affected by the number of days in advanced the appointment is scheduled
```

```python
data.groupby('lead_time')['Noshow'].count().plot(figsize=(15,7.5), color='b')
plt.show()
#shows the range of lead times and the how common a patient would schedule an appointment that far out
```

```python
lead_time_breakdown = pd.Series([
    'Same Day',
    '1-3 Days',
    '4-7 Days',
    '8-31 Days',
    '32+ Days'
])
data['leadtime_categories'] = pd.cut(data.lead_time, bins = [-1, 0, 3, 7, 31, 999], labels = lead_time_breakdown)

#I felt that there was a big variety of lead times and I wanted to be able to better analyze the affect of the 
#lead time so I broke them down into more manageable categories 
```

```python
ltc = data.leadtime_categories.value_counts()
ltc
```

```python
data.leadtime_categories.value_counts()

bottom_labels = data.leadtime_categories.unique()
y_pos = np.arange(5)
heights = ltc

plt.bar(y_pos, heights, align='center', alpha=0.5)
plt.xticks(y_pos, bottom_labels)
plt.ylabel('Number of Patients')
plt.title('Lead Time Frequencies')
 
plt.show()
#shows pattern for the amount of days ahead patients usually scheduled their appointments out
```

```python
lt_categ_noshow_grouped = data.groupby(['leadtime_categories', 'Noshow']).count()['PatientId']
lt_categ_noshow_grouped
#I grouped lead time categories with the amount of no-show patients in order to look for a pattern
```

```python
n_groups = 5
leadtime_shows = [36771, 11316, 13097, 20523, 6501]
leadtime_noshows = [1792, 3359, 4413, 9545, 3205]

#fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.9
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

rectsl = plt.bar(index, leadtime_shows, bar_width,
                color='b', label='Showed Up')
rects2 = plt.bar(index, leadtime_noshows, bar_width,
                color='r', label='No Show')

plt.xlabel('Appointment')
plt.ylabel('Number of Patients')
plt.title('Number of Days Appointment Scheduled Ahead vs. Show Rate')
plt.xticks(index, ('Same Day', '1-3 Days', '4-7 Days', '8-31 Days', '32+ Days'))
#show how many patients within each group of lead times would show up to their appointments vs. those who did not show
```

### We can see from the above chart that there's a higher probability a patient will not show for their appointment the longer ahead they scheduled the appointment.

```python
data['age_categories'] = data.Age.apply(lambda x: min(int(x / 10), 9))
#this is creating new column in our dataframe to reflect the age category of the patient so it's easier to analyze ages
#it will break the ages into 10 seperate age groups spanning 10 years each
#0: ages 0-9, 1: ages 10-19, 2: ages 20-29, 3: ages 30-39, 4: ages 40-49, 5: ages 50-59, 
#6: ages 60-69, 7: ages 70-79, 8: ages 80-89, 9: ages 90+
data.age_categories.value_counts()
```

```python
age_grouped = data.groupby(['age_categories', 'Noshow']).count()['PatientId']
age_grouped
#show the pattern of no-show rate based on the different age categories created
```

```python
i = 0 
age_show = []
age_noshow = []

while i < 10:
    show = age_grouped[i][0]
    age_show.append(show)
    no_show = age_grouped[i][1]
    age_noshow.append(no_show)
    i += 1
    
print(age_show, age_noshow)

#this code appends the count of patients who show to their appointments for each age group in one list and those
#who do not show to their appointment in another list
```

```python
n_groups = len(data.age_categories.unique())
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
 
rects1 = plt.bar(index, age_show, bar_width,
                 color='b',
                 label='Show')
 
rects2 = plt.bar(index + bar_width, age_noshow, bar_width,
                 alpha=0.5,
                 color='r',
                 label='Noshow')
 
plt.xlabel('Age Groups')
plt.ylabel('Number of Patients')
plt.title('Show Patterns for Age Groups')
plt.xticks(index + bar_width, (data.age_categories))
plt.legend()
 
plt.tight_layout()
plt.show()

#shows the number of patients who show vs. do not show to their appointment by age group in a visualization
```

```python
percent = age_noshow[1] / age_show[1]
percent
```

```python
i = 0
j = 0
age_ratios = []

while i < 10 and j < 10:
    percent = age_noshow[i] / age_show[j]
    age_ratios.append(float('%.3f'%(percent)))
    i += 1

print(age_ratios)
```

```python
barWidth = 0.85
labels = ('Ages 0-9', 'Ages 10-19', 'Ages 20-29', 'Age 30-39', 'Age 40-49',
        'Ages 50-59', 'Ages 60-69', 'Ages 70-79', 'Ages 80-89', 'Ages 90+')
y_pos = np.arange(len(labels))
performance = age_ratios

fig = plt.figure(figsize=(15,10))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.ylabel("Ratio of Noshow to Show")
plt.title('Percent of Age Groups that are No-shows')

plt.show()
```

#### From the chart above, we can see that each age group has a no show rate around 0.22 until we see the the size of that age group start to drop, we also see the rate of no-show drop. Overall though, the no-show rate is around 0.2.

```python
i = 0
age_percent = []

while i < 10:
    noshow = age_grouped[i][1]
    percent = round(noshow / total_patients*100, 2)
    #show = age_grouped[i][0]
    #percent = round(noshow / show*100, 2)
    age_percent.append(float('%.3f'%(percent)))
    i += 1

print(age_percent)
#this code finds the Noshow percent that that age group has overall compared to the total number of patients -
#this is unlike the last comparison, where I was looking at the rate of no show specifically within that age group
```

```python
barWidth = 0.85
labels = ('Ages 0-9', 'Ages 10-19', 'Ages 20-29', 'Age 30-39', 'Age 40-49',
        'Ages 50-59', 'Ages 60-69', 'Ages 70-79', 'Ages 80-89', 'Ages 90+')
y_pos = np.arange(len(labels))
performance = age_percent

fig = plt.figure(figsize=(15,10))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.ylabel("Percent of Age Group that's a No-show")
plt.title('Percent of Age Groups that are No-shows')

plt.show()
```

#### We can see here that the first half of the chart has higher rates of No-show appointments. We can assume that if a patient is 39 or younger, they will be more likely to not show up for their appointment.
